# Plotting an interesting building in 3D

To demonstrate this project, let's take an interesting building and look at the results. We take this building in the Cogels-Osylei in Antwerp. The addresses of this house group are p Cogels-Osylei 6-12, so we try number 6.

```python
from lidar import Address, Building
import matplotlib.pyplot as plt
```

We start by looking up the address with an Address object.

```python
example_address=Address(street="Cogels-Osylei", number="6", municipality="Antwerpen")
print(example_address)
```

To check if this is correct, we can plot the shape.

```python
example_address.get_building_shape().boundary.plot()
```

Now we want to put this in a `Building` instance. We set `auto_download` to `True` to show it off, but be aware that this will download two  +1GB files if you try to run this yourself. 

```python
example_building=Building(address=example_address,auto_download=True)
```

The Canopy Height Model is calculated, once the data is loaded. We can plot it as an image.

```python
fig,ax=example_building.plot_image("CHM", cmap="summer")
fig.set_size_inches(12,7)
```

This looks alright, so now let's go on to the main act: Making a 3D plot.

```python
%matplotlib nbagg
# to make plot interactive
fig,ax=example_building.plot3d("CHM", color="lightgrey")
fig.set_size_inches(12,12)
```
